```python
import numpy as np
```

```python
#цены за продажу
#нужно продать как можно больше 
prices = np.array([1, 3, 2, 4])
```

```python
# 1 нам сначала рандомно нужно попроходить до конечного состояния, записать те действия, что выполняем
# нам нужно пройтись обратно(теперь у нас есть количество шагов) и посчитать Q
```

```python
maxcust = np.array([6, 7, 4, 3])
maxvendor = 35
A = np.array([])
iter = 500
#ПРОХОД ВПЕРЕД
for i in range(iter):
  states = [0, 0, 0, 0]
  actions = ''
  last_actions = ''
  while True:
    action = np.random.choice([0, 1, 2, 3])                                       #выбрали действие
    if (states[action] + 1 < maxcust[action] and np.sum(states) < maxvendor):     #если удовлетворяет нашим границам, то тогда записываем что делали
      states[action] +=1
      last_actions = actions
      actions += str(action)
      actions += " "
    else: 
      last_actions = last_actions.strip() #удаляю пробелы в конце строк
      print("Последние состояния", states, 'Последние действия',last_actions)
      A = np.append(A, last_actions)
      break
A[:10]
```

```python
#ПРОХОД НАЗАД
gamma = 0.9
Q = np.zeros(len(A))
for i in range(len(A)):
  acts = A[i].split(" ")
  #print(acts)
  for act in reversed(acts):        # идем в обратном направлении
    # Функция Беллмана
    Q[i] = (-1 * prices[int(act)]) + gamma * Q[i]
Q[:5]
```

```python
A = A.reshape(iter,1)
A.shape
Q = Q.reshape(iter,1)
Q.shape
```

```python
QSA = np.append(Q, A, axis=1)
QSA[:16]
```

```python
print("Максимальный Q:", max(QSA[:, 0]))
ind = np.argmax(QSA[:,0])
print("Индекс", ind)

temp = QSA[ind]
stemp = temp[1].split(" ")
c0 = stemp.count("0")
c1 = stemp.count("1")
c2 = stemp.count("2")
c3 = stemp.count("3")
coefs = [c0, c1, c2, c3]
zatraty = np.dot(coefs, prices)
print(stemp)
print([c0, c1, c2, c3])
print("Затраты:", zatraty)
```

```python
print("Минимальный Q:", min(QSA[:, 0]))
ind = np.argmin(QSA[:,0])
print("Индекс", ind)

mtemp = QSA[ind]
mstemp = mtemp[1].split(" ")
c0 = mstemp.count("0")
c1 = mstemp.count("1")
c2 = mstemp.count("2")
c3 = mstemp.count("3")
coefs = [c0, c1, c2, c3]
zatraty = np.dot(coefs, prices)
print(mstemp)
print([c0, c1, c2, c3])
print("Затраты:", zatraty)
```
